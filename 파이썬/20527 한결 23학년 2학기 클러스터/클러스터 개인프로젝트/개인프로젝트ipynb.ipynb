```python

import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['figure.dpi'] = 100
import gdown
import pandas as pd
```

```python
import requests

url="http://data4library.kr/api/loanItemSrch?authKey=73d6409e91ab04b21c6fb98f6d5f46965694b5d9e27ba742ee910054e6ef5298&startDt=2023-08-01&endDt=2023-08-31&age=20&dtl_region=31070&format='JSON'"
r=requests.get(url)
data=r.json()
print(data)
```

```python
books=[]
for d in data['response']['docs']:
  books.append(d['doc'])
print(books)
```

```python
import pandas as pd

books_df=pd.DataFrame(books)
print(books_df)
```

```python
books=books_df[['no','ranking','bookname','authors','publisher','publication_year','isbn13','loan_count']]
books.head()
```

```python
books['loan_count']=books['loan_count'].astype(float)
```

```python
fig, ax=plt.subplots(figsize=(10,8))

sc=ax.scatter(books['publication_year'],books['publisher'],
           s=books['loan_count']**1.3,
           c=books["loan_count"])
ax.set_title("출판사별 발행도서")
fig.colorbar(sc)
fig.show()
```
