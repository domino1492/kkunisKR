```python
import gdown
url='https://drive.google.com/uc?id=1VpY5gjx_hcpWbk1KIvye5K5yjU0vUEpk'
filename='pt_book_fix.csv'
gdown.download(url, filename, quiet = False)
```

```python
import pandas as pd
pt_book=pd.read_csv(filename, low_memory=False)

import matplotlib.pyplot as plt
plt.scatter(pt_book['도서권수'],pt_book['대출건수'],alpha=0.1)
plt.show()
```

```python
%config InlineBackend.print_figure_kwargs={'bbox_inches':None}

plt.figure(figsize=(600/72,400/72))
#plt.figure(flgsize=(6,4))
plt.scatter(pt_book['도서권수'],pt_book['대출건수'],alpha=0.1)
plt.show()
```

```python
plt.scatter(pt_book['도서권수'],pt_book['대출건수'],alpha=0.1, marker='*', color="red") 
```

```python
fig,axs=plt.subplots(2, figsize=(6,8))

axs[0].scatter(pt_book['도서권수'],pt_book['대출건수'],alpha=0.1, marker='*', color="red")
axs[0].set_title('scatter plot')

axs[1].hist(pt_book['대출건수'],bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
```

```python
fig,axs=plt.subplots(1,2, figsize=(10,4))
axs[0].scatter(pt_book['도서권수'],pt_book['대출건수'],alpha=0.1, marker='*', color="red")
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

axs[1].hist(pt_book['대출건수'],bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')
```
