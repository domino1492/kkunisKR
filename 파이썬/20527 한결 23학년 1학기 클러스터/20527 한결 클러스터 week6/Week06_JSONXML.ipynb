JSON : JavaScript Object Notation

{
    "Key1" : "Value1"
    "Key2" : "Value2"
}

```python
d={"name" : "데이터 분석"}
print(d['name'])
print(type(d))
```

```python
import json
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
print(type(d_str))
```

```python
d2 = json.loads(d_str)
print(d2['name'])
print(type(d2))
```

```python
d3_str = '''
{
  "name" : "데이터 분석",
  "author" : ["홍길동", "김철수"],
  "year" : 2023
}
'''
d3 = json.loads(d3_str)
print(d3['author'][0])
```

```python
import pandas as pd
json_df = pd.read_json(d3_str)    #json -> DataFrame
list_df = pd.DataFrame(d3)       #Python 리스트 -> DataFrame

print(json_df)
print(type(json_df))
print(list_df)
print(type(list_df))
```
